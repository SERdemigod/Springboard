```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

### Q1: What is the trend of non-renewable energy sources in India from the 1960's to present day?

```python
#Isolate oil, gas, and coal
oil_gas_coal = energy[["Year", "Oil Consumption - EJ", "Gas Consumption - EJ", "Coal Consumption - EJ"]]
```

```python
#Put o,g,c into longform for seaborn use
lf_oil_gas_coal=oil_gas_coal.melt("Year", var_name = "Energy Source", value_name = "EJ")
```

```python
#Plot non-renewables
sns.lineplot(x="Year", y = "EJ", hue = "Energy Source", data = lf_oil_gas_coal)
```

### Q2: What is the general trend of renewable energy sources over this time period?

```python
#Isolate renewables
renewables = energy[["Year", 'Electricity from solar (TWh)', 'Electricity from other renewables (TWh)', 'Electricity from hydro (TWh)', 'Electricity from wind (TWh)']]
```

```python
#Put renewables into longform for seaborn use
lf_renewables=renewables.melt("Year", var_name = "Energy Source", value_name = "TWh")
```

```python
#Plot renewables
sns.lineplot(x="Year", y = "TWh", hue = "Energy Source", data = lf_renewables)
```

### Q3: How do renewable and fossil fuel based energy sources compare?

```python
#Convert renewables from TWh to EJ
energy['Electricity from solar (EJ)']= energy['Electricity from solar (TWh)']/278
energy['Electricity from other renewables (EJ)']= energy['Electricity from other renewables (TWh)']/278 
energy['Electricity from hydro (EJ)'] = energy['Electricity from hydro (TWh)']/278
energy['Electricity from wind (EJ)'] = energy['Electricity from wind (TWh)']/278
```

```python
renewables_EJ = energy[["Year", 'Electricity from solar (EJ)', 'Electricity from other renewables (EJ)', 'Electricity from hydro (EJ)', 'Electricity from wind (EJ)']]
```

```python
lf_renewables_EJ=renewables_EJ.melt("Year", var_name = "Energy Source", value_name = "EJ")
```

```python
#Plot renewables in Exajoules
sns.lineplot(x="Year", y = "EJ", hue = "Energy Source", data = lf_renewables_EJ)
```

```python
#Plot renewables next to non-renewables in EJs
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize = (12,5))
sns.lineplot(x="Year", y = "EJ", hue = "Energy Source", data = lf_renewables_EJ, ax=ax2)
sns.lineplot(x="Year", y = "EJ", hue = "Energy Source", data = lf_oil_gas_coal, ax=ax1)
```

```python
#Bad news: When we plot against the same y-axis values we can see the extreme difference in scale.
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize = (12,5), sharey=True)
sns.lineplot(x="Year", y = "EJ", hue = "Energy Source", data = lf_renewables_EJ, ax=ax2)
sns.lineplot(x="Year", y = "EJ", hue = "Energy Source", data = lf_oil_gas_coal, ax=ax1)
```

### Q4: Is there a reason for the noticeable increase in renewables starting in the early 2000's?

Good news(?): "Electricity Act 2003 was the first comprehensive framework that spurred development of renewable power in the country unfolding a developing regulatory structure composed of preferential tariffs, renewable purchase obligation and leading over to facilitating renewable energy certificates."

Source: https://www.iitr.ac.in/wfw/web_ua_water_for_welfare/water/WRDM/mnre_paper_Renew_Energy_2010.pdf

```python
#Noticeable improvements have been made since governmental regulations and incentives were put in place 
plt.axvline(2003, c = 'black')
sns.lineplot(x="Year", y = "EJ", hue = "Energy Source", data = lf_renewables_EJ)
```

### Q5: What other metrics are in the data that might be useful to track?

```python
#Metric to try to decline moving forward
sns.lineplot(x="Year", y="Annual CO2 emissions per unit energy (kg per kilowatt-hour)", data = energy)
```

```python
#Metric to try to decline moving forward
plt.figure(figsize=(22, 10))
sns.set(font_scale = 2)
p= sns.barplot(x="Year", y="Annual change primary energy consumption (%)", data = energy)
p.set_ylabel("Change in Energy Consumption(%)", fontsize = 30)
for ind, label in enumerate(p.get_xticklabels()):
    if ind % 10 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```

### We can see a downward trend in both Annual CO2 emissions per unit energy and Annual change primary energy consumption (%). However, these trends are minor and we would hope to see more drastic reductions going forward
